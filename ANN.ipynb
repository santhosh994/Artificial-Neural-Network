```python
#import dataset
dataset = pd.read_csv("C:\\Users\\santy\\Desktop\\Krish Naik DS\\ANN\\Complete-Deep-Learning-master\\Complete-Deep-Learning-master\\ANN\\Churn_Modelling.csv")
```

```python
#exited is the target variable/dependent feature
#columns rownumber and surname are non-essential 
dataset.head()
```

```python
x = dataset.iloc[:,3:13]
y = dataset.iloc[:,13]
```

```python
x.head()
```

```python
y.head()
```

```python
#creating dummy variables for Geography and Gender
geography = pd.get_dummies(x['Geography'], drop_first=True)
gender = pd.get_dummies(x['Gender'], drop_first=True)
```

```python
#concantenate
x = pd.concat([x,geography,gender],axis=1)
```

```python
#drop old categorical columns
x=x.drop(['Geography','Gender'], axis=1)
```

```python
x.head()
```

```python
#split into train and test data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state= 0)
```

```python
#feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
#importing keras libraries
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LeakyReLU,PReLU,ELU
from keras.layers import Dropout
```

```python
#initialising the ANN
classifier = Sequential()
```

```python
#adding first input layer and first hidden layer
#units is the output dimensions, input_dim is no of input features
classifier.add(Dense(units=6, kernel_initializer='he_uniform', activation='relu', input_dim=11))
```

```python
#adding second hidden layer
classifier.add(Dense(units=6, kernel_initializer='he_uniform', activation='relu'))
```

```python
#adding output layer
classifier.add(Dense(units=1, kernel_initializer='glorot_uniform', activation = 'sigmoid'))
```

```python
classifier.summary()
```

```python
#compiling the ANN
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
#fitting the ANN to training dataset
model_history = classifier.fit(x_train, y_train, validation_split=0.33, batch_size=10, epochs=100)
```

```python
#predicting the Test set results
y_pred = classifier.predict(x_test)
y_pred = (y_pred>0.5)
```

```python
#confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
```

```python
cm
```

```python
#accuracy
from sklearn.metrics import accuracy_score
score = accuracy_score(y_pred, y_test)
```
